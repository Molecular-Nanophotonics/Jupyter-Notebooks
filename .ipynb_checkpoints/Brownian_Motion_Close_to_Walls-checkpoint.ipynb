# Brownian Motion Close to Walls

1. Lobry, L. *et al.*, Diffusion of Brownian ParticlesTrapped Between Two Walls: Theory and Dynamic-Light-Scattering Measurements, *Phys. Rev. B* **53**, 12050 (1996) https://doi.org/10.1103/PhysRevB.53.12050

```python
import numpy as np
pi = np.pi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams.update({'font.size' : 20,
                     'axes.labelpad' : 10,
                     'axes.linewidth': 1.0,
                     'xtick.top' : True,
                     'xtick.direction' : 'in',
                     'ytick.right' : True,
                     'ytick.direction' : 'in',
                     'lines.linewidth' : 2,
                     'lines.markersize': 3})
```

```python
def f1(x):
    return (1 - 9/16*(1/x) + 1/8*(1/x)**3 - 45/256*(1/x)**4 - 1/16*(1/x)**5)**(-1)

def f2(xlist):
    nlist0 = np.arange(0,100)
    nlist1 = np.arange(1,100)
    return np.sum([f1((2*nlist0+1)*x) for x in xlist], axis=1) + np.sum([f1((2*nlist1-1)*x) for x in xlist], axis=1) - 2*np.sum([f1(2*nlist1*x) for x in xlist], axis=1)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))

x = np.linspace(1, 5, 100)
ax.plot(x, 1/f1(x), 'b', label='1 Wall')
ax.plot(x, 1/f2(x), 'r', label='2 Walls')
ax.set_ylim(0, 1)
ax.set_xlim(x.min(), x.max())
ax.set_ylabel('$D_\parallel/D$')
ax.set_xlabel('$H/(2R)$')
ax.legend(loc='center right')
plt.savefig('Brownian_Motion_Close_to_Walls.pdf',bbox_inches='tight',pad_inches=0.025);
plt.show()
```
